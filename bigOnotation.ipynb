```python
# Setup all the dependencies

import matplotlib.pyplot as plt
import numpy as np
import timeit

plt.rcParams['figure.figsize'] = [10, 6] # set size of plot
```

```python
## Creating the function to sum all elements in a given array


def find_sum(given_array):
  total = 0
  for i in given_array:
    total += i
  return total


  ## Plotting the values


ns = np.linspace(10, 100, 10, dtype=int)
ts = [timeit.timeit('find_sum(lst)',
                    setup='lst=list(range({})); random.shuffle(lst)'.format(n),
                    globals=globals(),
                    number=100)
      for n in ns]
plt.plot(ns, ts, 'or')

# line of best fit for red plots
degree = 4
coeffs = np.polyfit(ns, ts, degree)
p = np.poly1d(coeffs)
plt.plot(ns, [p(n) for n in ns], '-r')
```
