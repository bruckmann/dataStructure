```python
# Setup all the dependencies

import matplotlib.pyplot as plt
import numpy as np
import timeit

plt.rcParams['figure.figsize'] = [10, 6] # set size of plot
```

```python
## Creating a function to sum all elements in a given array - Linear time - O(n)


def find_sum(given_array):
  total = 0
  for i in given_array:
    total += i
  return total


  ## Plotting the values


ns = np.linspace(10, 100, 10, dtype=int)
ts = [timeit.timeit('find_sum(lst)',
                    setup='lst=list(range({})); random.shuffle(lst)'.format(n),
                    globals=globals(),
                    number=100)
      for n in ns]
plt.plot(ns, ts, 'or')

# line of best fit for red plots
degree = 4
coeffs = np.polyfit(ns, ts, degree)
p = np.poly1d(coeffs)
plt.plot(ns, [p(n) for n in ns], '-r')
```

In this example, the algorithm exhibits **linear** time complexity, meaning its **runtime scales directly with the input size**. Using Big O notation, we express this as **O(n)**, where n represents the number of elements in the input array.

```python
## Retrieving a element in an array for a given index  - Constant time - O(1)

lst = list(range(1_000_000))
ns = np.linspace(0, len(lst), 1000, endpoint=False, dtype=int)
ts = [timeit.timeit('_ = lst[{}]'.format(n),
                    globals=globals(),
                    number=10000)
for n in ns]

plt.plot(ns, ts,'or')

degree = 4
coeffs = np.polyfit(ns, ts, degree)
p = np.poly1d(coeffs)
plt.plot(ns, [p(n)for n in ns],'-b')
```

In this example, the algorithm exhibits **constant time complexity**. This means that the runtime is independent of the input size, regardless of how large the array becomes. Using Big O notation, we express this as **O(1)**, where n represents the number of elements in the input array.

```python
## Fibonacci function - Quadratic time - o(n^2)

def fib(n):
   if n <= 1:
       return n
   else:
       return fib(n-1) + fib(n-2)

ns = range(5, 30)
ts = [timeit.timeit('fib({})'.format(n),
                    globals=globals(),
                    number=1)
         for n in ns]
plt.plot(ns, ts, 'or')

degree = 4
coeffs = np.polyfit(ns, ts, degree)
p = np.poly1d(coeffs)
plt.plot(ns, [p(n)for n in ns],'-b')
```

The Fibonacci algorithm exhibits **quadratic time complexity**, which is expressed as **O(n^2)**. This means that the runtime increases as the square of the input size. For example, if the input size is 10, the runtime is 100. If the input size is 100, the runtime is 10,000.
